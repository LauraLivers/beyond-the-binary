<h2>Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.patheffects as PathEffects
```

<h2>Text Properties

```python
font_title = {'fontname' : 'din condensed', 'color' : 'darkgrey', 'size':'70', 'weight':'bold'}
font_anno_dark = {'fontname' : 'din condensed', 'color' : 'darkgrey', 'size':'25'}
font_anno_dark_small = {'fontname' : 'din condensed', 'color' : 'darkgrey', 'size':'15'}
font_anno_light = {'fontname' : 'din condensed', 'color' : 'oldlace', 'size' : '35'}
font_anno_light_small = {'fontname' : 'din condensed', 'color' : 'oldlace', 'size' : '25'}
font_background = {'fontname' : 'din condensed', 'color' : 'white', 'size' : '135'}
font_sankey = {'fontname' : 'din condensed', 'color' : 'white', 'size' : '25', 
               'path_effects': [PathEffects.withStroke(linewidth=3, foreground='b')]}
rect_prop = {'color': '#79F10E', 'alpha': 0.2, 'zorder': 1}
```

<h2>Plot

```python
countries = ['NL', 'CH', 'EE']
y_pos = np.arange(len(countries))[::-1]
bar_width = 0.6

# PLOT
fig, ax = plt.subplots(figsize=(12, 4))

ax.barh(y_pos, [28] * len(countries), color='mediumblue', height=bar_width, edgecolor='none')

values = [0, 1, 2]
ax.barh(y_pos, values, color='deeppink', height=bar_width, edgecolor='none')

ax.set_yticks(y_pos)
ax.set_yticklabels(countries, **font_anno_dark)

ax.set_title('Computer Science in School', **font_title, x=0.42, y=1.3)
text_title_anno=ax.text(-0.9, 2.5, '''In light of recent technological advances teachers are advised to use 
technology in all subjects, this chart only represents the time specifically 
allocated for school children to learn computer literacy.'''
                        , **font_anno_dark_small)


text_anno = ax.text(0.33, 0.85, '1', 
                    **font_anno_light)
text_anno.set_path_effects([PathEffects.withStroke(linewidth=1, foreground='mediumblue')])
text_anno_detail = ax.text(11.1, 0.9, '45\' in Comuter & Media per week in grade 4', **font_anno_light_small)

text_anno2 = ax.text(0.33, -0.15, '2', 
                    **font_anno_light)
text_anno2.set_path_effects([PathEffects.withStroke(linewidth=1, foreground='mediumblue')])
text_anno2_detail = ax.text(11.1, -0.15, '60\' of Computer Science per week in grade 4', **font_anno_light_small)

text_anno3 = ax.text(11.1, 1.9, '  0\' allocated for Computer Science', **font_anno_light_small)


ax.set_xlim(0, 30);
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.tick_params(left=False, bottom=False, labelleft=True, labelbottom=False)
```
